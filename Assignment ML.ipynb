```python
dataset=pd.DataFrame(boston_df.data,columns=boston_df.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price']=boston_df.target
```

```python
# stats of data
dataset.describe()
```

```python
# Missing value
dataset.isnull()
```

```python
# EDA
# correlation  
dataset.corr()
```

```python
plt.scatter(dataset['RM'],dataset['Price'])
plt.xlabel("RM")
plt.ylabel("Price")
```
